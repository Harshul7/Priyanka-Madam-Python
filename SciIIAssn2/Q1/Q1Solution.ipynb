### Question 1a

```python
I1=1
randomNumbers = [I1]
A=106
C=1283
M=6075
last=1
for j in range(999):
    last=(A*last+C)%M 
    randomNumbers.append(last)

print(randomNumbers)
```

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.array(randomNumbers)
plt.plot(range(1,1001),array[0:1000])
plt.xlabel("j")
plt.ylabel("$I_(j)$")
```

### Question 1c


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
A=106
C=1283
M=6075
I1=1

N_values=[1,10,50]+list(range(100,2100,100))
max_N = 2000
I_Sequence=[I1]
for _ in range(1,max_N):
    next_I=(A*I_Sequence[-1]+C)%M
    I_Sequence.append(next_I)

E_values=[]
for N in N_values:
    E_N=np.mean(I_Sequence[:N])
    E_values.append(E_N)

plt.plot(N_values,E_values)
plt.xlabel(r"$N$")
plt.ylabel(r"$E(I_j)$")
plt.title(r'Expectation value $E(I_j)$ vs $N$')

```
