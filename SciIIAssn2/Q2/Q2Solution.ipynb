#### Problem 1
- $ f(x) $ = $ x{\sin x}$
- a.

```python
import numpy as np
import matplotlib.pyplot as plt

a=-np.pi
b=np.pi 

N=100000

x=np.random.uniform(a,b,N)

print(x)

def f(x):
    return x*np.sin(x)

y=(b-a)*np.mean(f(x))

print(y)
```

- b.

```python
from scipy.integrate import quad
exact_value,_=quad(f,a,b)
print(exact_value)
# Plot v. N

N_values = list(range(1,100001,1000))
integral_estimates=[]

for N in N_values:
    x_random_temp=np.random.uniform(a,b,N)
    integral_estimate_temp=(b-a)*np.mean(f(x_random_temp))
    integral_estimates.append(integral_estimate_temp)

plt.plot(N_values,integral_estimates)
plt.xlabel(r"$N$")
plt.ylabel(r"Estimated Integral")
plt.title(r"Monte Carlo Integration of $x{\sin x}$ vs $N$")
plt.show()
```

```python
import numpy as np

#Constants
a= -np.pi/2
b=np.pi/2

N=100000

x=np.random.uniform(a,b,N)

def f(x):
    return np.cos(x)-(np.cos(x))**2
#return np.cos(x)-np.cos(x)*np.cos(x)

print(x)

y=(b-a)*np.mean(f(x))

print(f"Monte Carlo Integration of cos x - cos^2(x)=",y)


```

```python
import matplotlib.pyplot as plt
from scipy.integrate import quad 

actual_value,_=quad(f,a,b)
print("actual_value=",actual_value)

N_values=list(range(1,100001,1000))
integral_estimates=[]

for N in N_values:
    x_random_temp=np.random.uniform(a,b,N)
    integral_estimate_temp=(b-a)*np.mean(f(x_random_temp))
    integral_estimates.append(integral_estimate_temp)

plt.plot(N_values,integral_estimates)
plt.xlabel(r"$N$")
plt.ylabel(r"Estimated Integral")
plt.title(r"Monte Carlo Integration of ${\cos x}-{\cos^2(x)}$ vs $N$")
plt.show()
```
