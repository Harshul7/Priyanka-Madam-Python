#### Drunkard Returning to Origin Problem
- Find Probability
- Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt


def fact(N):
    if (N==0):
        return 1
    return N*fact(N-1)
    
probabilityAtOrigin = []
N_values=list(range(1,101))
for N in N_values:
    if(N%2!=0):
        probabilityAtOrigin.append(0)
    else:
        probabilityAtOrigin.append(fact(N)/(fact(N/2)*fact(N/2))/2**N)
print(probabilityAtOrigin)
plt.plot(N_values,probabilityAtOrigin)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability back at origin after N steps")
plt.title(r"Drunkard Random Walk Plot")
plt.show()
```
