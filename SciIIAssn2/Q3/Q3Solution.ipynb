```python
import numpy as np
import matplotlib.pyplot as plt


def fact(N):
    if (N==0):
        return 1
    return N*fact(N-1)
    
probabilityAtOrigin = []
N_values=list(range(1,101))
for N in N_values:
    if(N%2!=0 and N>=3):
        probabilityAtOrigin.append(fact(N)/(fact((N-3)/2)*fact((N+3)/2))/(2**N))
    elif(N%2==0 and N>=3):
        probabilityAtOrigin.append(0)
    elif(N<3):
        probabilityAtOrigin.append(0)
print(probabilityAtOrigin)
plt.plot(N_values,probabilityAtOrigin)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability back at origin after N steps")
plt.title(r"Drunkard Random Walk Plot")
plt.show()
```

- The correct probability = $ \frac{\binom{N}{(N-3)/2}}{2^N}$

```python
import numpy as np
import matplotlib.pyplot as plt
#Including num_trials=10000 
# Number of random walk simulations for each n
num_trials = 10000
N_values = list(range(1,101))
initial_position = a= -3 

probabilityAtOrigin=[]

for N in N_values:
    steps=np.random.choice([-1,1],size=(num_trials,N))
    final_positions=initial_position+np.sum(steps,axis=1)

    # Count how many times person returned to origin
    num_returns=np.sum(final_positions==0)
    prob_return=num_returns/num_trials 

    probabilityAtOrigin.append(prob_return)

plt.plot(N_values,probabilityAtOrigin)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability of return to Origin")
plt.title(r"$1D Random Walk$")
#plt.figure(figSize=(10,6))
    
```
