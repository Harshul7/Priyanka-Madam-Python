#### Drunkard Returning to Origin Problem
- Find Probability
- Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt


def fact(N):
    if (N==0):
        return 1
    return N*fact(N-1)
    
probabilityAtOrigin = []
N_values=list(range(1,101))
for N in N_values:
    if(N%2!=0 and N>=3):
        probabilityAtOrigin.append(fact(N)/(fact((N-3)/2)*fact((N+3)/2))/(2**N))
    elif(N%2==0 and N>=3):
        probabilityAtOrigin.append(0)
    elif(N<3):
        probabilityAtOrigin.append(0)
print(probabilityAtOrigin)
plt.plot(N_values,probabilityAtOrigin)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability back at origin after N steps")
plt.title(r"Drunkard Random Walk Plot")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Including num_trials=10000 
# Number of random walk simulations for each N
num_trials = 10000
N_values = list(range(1,101))
initial_position = a= -3 

probabilityAtOrigin=[]

for N in N_values:
    steps=np.random.choice([-1,1],size=(num_trials,N))
    final_positions=initial_position+np.sum(steps,axis=1)

    # Count how many times person returned to origin
    num_returns=np.sum(final_positions==0)
    prob_return=num_returns/num_trials 

    probabilityAtOrigin.append(prob_return)

plt.plot(N_values,probabilityAtOrigin)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability of return to Origin")
plt.title(r"$1D Random Walk$")
#plt.figure(figSize=(10,6))
    
```

```python
import numpy as np
import matplotlib.pyplot as plt 
initial_position_a= -3 
initial_position_b= 3 

num_trials = 10000
N_values=list(range(1,101))

probabilityOfMeeting = []
for N in N_values:
    steps_a=np.random.choice([-1,1],size=(num_trials,N))
    steps_b=np.random.choice([-1,1],size=(num_trials,N))
    final_positions_a=initial_position_a + np.sum(steps_a,axis=1)
    final_positions_b=initial_position_b+np.sum(steps_b,axis=1)

    num_meet_again=np.sum(final_positions_a==final_positions_b)
    prob_meet_again=num_meet_again/num_trials

    probabilityOfMeeting.append(prob_meet_again)

plt.plot(N_values,probabilityOfMeeting)
plt.xlabel(r"$N$")
plt.ylabel(r"Probability of the 2 drunkards meeting")
plt.title(r"")
```

```python
import numpy as np
import matplotlib.pyplot as plt

initial_position=0

N_values=list(range(1,101))
num_trials=10000

meanDisplacement =[]

for N in N_values:
    steps=np.random.choice([-1,1],size=(num_trials,N))
    final_positions=initial_position+np.sum(steps,axis=1)

    sum=0
    for pos in final_positions:
        sum+=abs(pos)
    sum/=num_trials
    meanDisplacement.append(sum)

plt.plot(N_values,meanDisplacement)
plt.xlabel(r"$N$")
plt.ylabel(r"mean Displacement")
plt.title(r"mean Displacement vs $N$")
plt.show()
```
